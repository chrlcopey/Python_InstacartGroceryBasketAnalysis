```python
df = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'df_all_merge.pkl'))
```

# Histogram

```python
# creating histogram for 'order_hour_of_day'
hist = df['order_hour_of_day'].plot.hist(bins = 24)
```

The histogram makes it easy to see which hours of the day have the highest sales volume. The highest sales volume occurs between 10:00 and 15:00.

# Bar Chart - Distribution of orders amount customers

```python
# creating a bar chart for "loyalty_flag"
bar = df['loyalty_flag'].value_counts().plot.bar()
```

# Line Chart - difference in expenditure depending on the hour of the day

```python
#creating a sample of the dataframe
sampled_df = df.sample(n=100000, random_state=1)
```

```python
#group by hour of the day and calculate the mean expenditure
hourly_expenditure = sampled_df.groupby('order_hour_of_day')['prices'].mean().reset_index()
```

```python
#creating a line chart
line_price_hod = sns.lineplot(data = sampled_df, x = 'order_hour_of_day', y = 'prices')
```

# Line chart - Connections between age and number of dependents

```python
sampled_df.columns
```

```python
#creating a line chart
line_age_nod = sns.lineplot(data = sampled_df, x = 'age', y = 'number_dependents')
```

There is not clear connecton between age and number of dependents.

# Scatterplot - connection between age and spending power (income)

```python
#creating a scatterplot of 'age' and 'income'
scatter = sns.scatterplot(x='age', y='income', data=sampled_df)
```

There is a clear connection between the age and spending power within people 40 years or older.
